```python
## K: carry capacity(maximum sustainable population)
## N: start population 
## r: rate of change(maximum population growth)

## population growth model 

## logistic euqation(continuous):
## dN/dt = r*N*(1-K/N) 
## x = N/K
## dN/dt = r*x*(1-x)
## X(t): sigmoid function 

## difference equation: 
## X(n+1) = r*X(n) yearly growth

## logistic map:(discrete):
## logistic differential equation: analogous to logistic equation 
## x(n+1) = r*x(n)*(1-x(n))
## r[0:4], x[0:1]
```

```python
import numpy as np
R = np.linspace(0, 4, 80)
x = 0.5 
iter_num = 65
```

```python
import matplotlib.pyplot as plt 
from matplotlib.animation import FuncAnimation 
plt.rcParams['animation.html']='html5'

fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
line, = ax.plot((0,65),(0,1))

def animation(i):
    r = R[i]
    l = f'r = {r:.2f}'
    data = x
    datas = []
    for iteration in range(iter_num):
        datas.append(data)
        data = r * data * (1-data)
    line.set_data(range(iter_num), datas)
    line.set_label(l)
    ax.legend()
    return line, 
    
anim = FuncAnimation(fig, animation, frames=len(R),
                    interval=100, blit=True)
anim.save('img/logistic_difference_equation.gif', writer='pillow')
plt.close()
anim
```
